```python
from numpy import cov, linalg, sqrt, ones, diag
```

```python
from matplotlib import pyplot
```

```python
def principalComponentAnalysis (data, n=2):
    samples, features = data.shape
    meanVec = data.mean(axis = 0)
    dataC = (data - meanVec).T
    covar = cov(dataC)
    evals, evecs = linalg.eig(covar)
    indices = evals.argsort()[::-1]
    evecs = evecs[:,indices]
    basis = evecs[:,:n]
    energy = evals[:n].sum()
    # norm wrt to variance
    # sd = sqrt (diag(covar))
    # zscores = dataC.T /s d
    
    return basis, energy
```

```python
def principalComponentAnalysis (data, n=2):
    samples, features = data.shape
    meanVec = data.mean(axis = 0)
    dataC = (data - meanVec).T
    covar = np.cov(dataC)
    evals, evecs = np.linalg.eig(covar)
    indices = evals.argsort()[::-1]
    evecs = evecs[:,indices]
    basis = evecs[:,:n]
    energy = evals[:n].sum()
    # norm wrt to variance
    # sd = sqrt (diag(covar))
    # zscores = dataC.T /s d
    
    return basis, energy
```

```python
from numpy import random, dot, array, outer
import numpy as np
```

```python
def extractPrincipaComponent(data, precision = 1e-9):
    samples, features = data.shape
    meanVec = data.mean(axis = 0)
    dataC = data - meanVec
    pc1 = random.random(features)
    pc0 = pc1 - 1.0
    while abs((pc0-pc1).sum()) > precision:
        t = np.zeros(features)
        for datum in dataC:
            t +=dot(datum, pc1) * datum
        pc0 = pc1
        pc1 = t /sqrt(dot (t,t))
    return pc1
```

```python
testData =  random.normal(0.0, 2.0, (100,2))
shear = array ([[2,1], [1,0]])
testData = dot(testData, shear)
pc1 = extractPrincipaComponent(testData)
print('Quick PC1:', pc1)
basis, energy = principalComponentAnalysis(testData, n=2)
print('Full PCA', "\n",basis, "\n",energy)
```
